Here is where we will use OpenCV to do some image analysis.  We will use haarcascades

```
# Standard imports
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import os

#setup our haar cascade
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

#Setup the file director
DIR = "OBJTest_resized"
OutDIR = "peopleOut"

#loop through every iamage in the folder and do this stuff to i
imgFiles = [f for f in os.listdir(DIR) if f.endswith('.png')]
for imgName in imgFiles:
	
	fName = os.path.join(DIR, imgName)
	# Read image
	img = cv2.imread(fName)
	im = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

	
	people, weights = hog.detectMultiScale(im, winStride=(1,16))
	print(people)
	for (x,y,w,h) in people:
		img = cv2.rectangle(img, (x, y), (x+w, y+h),(255,0,0),5)#BGR, 2 is stroke weight

	outName = os.path.join(OutDIR, imgName)
	cv2.imwrite(outName,img)
	
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```
import numpy as np 
import cv2
from google.colab.patches import cv2_imshow
import os


DIR = 'OBJTest_resized'
OutDIR = 'peopleMask'
#loop through every iamage in the folder and do this stuff to i
imgFiles = [f for f in os.listdir(DIR) if f.endswith('.png')]
for fName in imgFiles:
  print(fName)
  
  image = cv2.imread(os.path.join(DIR,fName))
  mask = np.zeros(image.shape, dtype=np.uint8)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  thresh = cv2.threshold(gray, 100, 200, cv2.THRESH_BINARY )[1]

  cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  cnts = cnts[0] if len(cnts) == 2 else cnts[1]
  threshold_area = 500
  for c in cnts:
      area = cv2.contourArea(c)
      if area > threshold_area:
          cv2.drawContours(mask, [c], -1, (255,255,255), -1)

  
  outName = os.path.join(OutDIR, fName)
  cv2.imwrite(outName,mask)
cv2.waitKey()
```
