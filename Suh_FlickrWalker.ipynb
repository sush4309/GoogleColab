```
import flickrapi
import urllib.request
import PIL.Image
#import png
import shutil

import datetime
import calendar
import os
import numpy as np

def resize_and_crop(img, size):


    # Get current and desired ratio for the images
    img_ratio = img.size[0] / float(img.size[1])
    ratio = size[0] / float(size[1])
    #The image is scaled/cropped vertically or horizontally depending on the ratio
    if ratio > img_ratio:
        img = img.resize((size[0], int(size[0] * img.size[1] / img.size[0])), PIL.Image.ANTIALIAS)
        box = (0, (img.size[1] - size[1]) / 2, img.size[0], (img.size[1] + size[1]) / 2)
        img = img.crop(box)
    elif ratio < img_ratio:
        img = img.resize((int(size[1] * img.size[0] / img.size[1]), size[1]), PIL.Image.ANTIALIAS)
        box = ((img.size[0] - size[0]) / 2, 0, (img.size[0] + size[0]) / 2, img.size[1])
        img = img.crop(box)
    else :
        img = img.resize((size[0], size[1]),PIL.Image.ANTIALIAS)
        # If the scale is the same, we do not need to crop
    return img
#min_upload_date=yesterday,
#max_upload_date=today,
def main():
    
    # Flickr api access key
    flickr = flickrapi.FlickrAPI('30c47280bf3821edb3f838e269d64ce2', '0b80cba0042b2a92', cache=True)
    # 'yesterday' is the oldest date, 'today' is the newest date
    yesterday = datetime.date.today() - datetime.timedelta(356)
    today = datetime.date.today() - datetime.timedelta(1)
    print(yesterday)
    unix_time=calendar.timegm(yesterday.timetuple())
    print(unix_time)
    unix_time2 = calendar.timegm(today.timetuple())
    print(unix_time2)
    photos = flickr.walk(lat= 34.04533472024815,
                         lon= -118.25074608649481,
                         radius=3,
                         
                         tag_mode = 'any',
                         tags='wall,building',
                         extras='url_c',
                         per_page=1000,  # may be you can try different numbers..
                         sort='interestingness-desc')

    urls = []
    for i, photo in enumerate(photos):
        print (i)
        url = photo.get('url_c')
        t = "{:04d}".format(i)
        fStr = 'wallNfacade'
        fName = fStr + '_' + t +'.png'
        if type(url) is str:
            urllib.request.urlretrieve(url,fName)
            # Resize the image and overwrite it
            image = PIL.Image.open(fName)
            image = resize_and_crop(image,[1024,1024])
            image = image.convert('RGB')
            image.save(fName)
            #shutil.move( fName, os.path.join(DIR,fName))
            if i > 1000:
                break
        else:
            break
            
if __name__ == "__main__":
    main()
```
